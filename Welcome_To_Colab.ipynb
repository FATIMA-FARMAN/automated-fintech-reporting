```
# ========================
# 2. Upload and unzip ZIP
# ========================
from google.colab import files
import zipfile, os, pandas as pd

uploaded = files.upload()  # upload archive (9).zip
zip_name = list(uploaded.keys())[0]   # detect uploaded filename
extract_to = "/content/data"

with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("‚úÖ Files extracted:")
for root, dirs, files in os.walk(extract_to):
    for file in files:
        print(os.path.join(root, file))
```

```
# ========================
# 3. Auto-detect CSV
# ========================
csv_file = None
for root, dirs, files in os.walk(extract_to):
    for file in files:
        if file.endswith(".csv"):
            csv_file = os.path.join(root, file)
            break

if csv_file:
    print("üìÇ Found CSV:", csv_file)
    df = pd.read_csv(csv_file)
    print("‚úÖ CSV loaded:", df.shape)
else:
    raise FileNotFoundError("‚ùå No CSV file found in the archive.")
```

```
# ========================
# 4. Connect to MySQL
# ========================
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="rootpass",
    database="creditcard_db"
)
cursor = conn.cursor()
print("‚úÖ Connected to MySQL")
```

```
# ========================
# 5. Create transactions table
# ========================
cursor.execute("""
CREATE TABLE IF NOT EXISTS transactions (
    transaction_id BIGINT AUTO_INCREMENT PRIMARY KEY,
    time INT,
    v1 FLOAT, v2 FLOAT, v3 FLOAT, v4 FLOAT, v5 FLOAT,
    v6 FLOAT, v7 FLOAT, v8 FLOAT, v9 FLOAT, v10 FLOAT,
    v11 FLOAT, v12 FLOAT, v13 FLOAT, v14 FLOAT, v15 FLOAT,
    v16 FLOAT, v17 FLOAT, v18 FLOAT, v19 FLOAT, v20 FLOAT,
    v21 FLOAT, v22 FLOAT, v23 FLOAT, v24 FLOAT, v25 FLOAT,
    v26 FLOAT, v27 FLOAT, v28 FLOAT,
    amount DECIMAL(10,2),
    class TINYINT
);
""")
print("‚úÖ Table created")
```

```
# ========================
# 6. Insert CSV into MySQL
# ========================
rows_inserted = 0
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO transactions (
            time, v1, v2, v3, v4, v5, v6, v7, v8, v9,
            v10, v11, v12, v13, v14, v15, v16, v17, v18,
            v19, v20, v21, v22, v23, v24, v25, v26, v27, v28,
            amount, class
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
                  %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
                  %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))
    rows_inserted += 1
    if rows_inserted % 10000 == 0:
        conn.commit()
        print(f"Inserted {rows_inserted} rows...")

conn.commit()
print(f"üìä All data inserted successfully! Total rows: {rows_inserted}")
```

```
# ========================
# 7. Fraud summary report
# ========================
def fraud_summary():
    query = """
    SELECT class, COUNT(*) as count, AVG(amount) as avg_amount
    FROM transactions
    GROUP BY class
    """
    df_summary = pd.read_sql(query, conn)
    df_summary.to_csv("/content/fraud_summary.csv", index=False)
    print("üìÇ Fraud report saved as fraud_summary.csv")
    return df_summary

fraud_summary()
```

```
import time

def scheduled_reports(interval_minutes=60, runs=3):
    """
    Generate fraud summary every X minutes for a given number of runs.
    interval_minutes: gap between reports (default 60 minutes)
    runs: how many times to run (default 3)
    """
    for i in range(runs):
        print(f"\n‚è≥ Run {i+1} of {runs} ‚Äî Generating report...")
        df_report = fraud_summary()
        display(df_report)
        if i < runs - 1:
            print(f"‚úÖ Sleeping for {interval_minutes} minutes...\n")
            time.sleep(interval_minutes * 60)

# Example: run 3 times, every 5 minutes (for testing)
scheduled_reports(interval_minutes=5, runs=3)
```

```
scheduled_reports(interval_minutes=60, runs=24)  # hourly for 1 day
```
